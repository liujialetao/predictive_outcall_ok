```python
import pandas as pd
import os
```

```python
data1 = pd.read_csv('solution1_2000_test1.csv', index_col=0)
data3 = pd.read_csv('solution3_2000_test1.csv', index_col=0)
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

x1 = data1.loss_ration
y1 = data1.average_idle_time
x3 = data3.loss_ration
y3 = data3.average_idle_time

plt.scatter(x1, y1, s=20, c='green', marker='*',label='solution1')
plt.scatter(x3, y3, s=20, c='red', marker='.',label='solution3')
plt.legend()
plt.title('result')
plt.xlabel('call_loss_ration')
plt.ylabel('wait_time')
plt.show()
```
